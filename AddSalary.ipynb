```python
import pandas as pd
```

```python
df1 = pd.read_csv('newdat/games_extended_with_dist.csv')
```

```python
df1.columns
```

```python
date_dict = {}
```

```python
import datetime
```

```python
df1['GAME_DATE_EST'] = pd.to_datetime(df1['GAME_DATE_EST'])
```

```python
df1['GAME_DAY'] = df1['GAME_DATE_EST'].dt.dayofweek
```

```python
df1['DAY_OF_YEAR'] = df1['GAME_DATE_EST'].dt.dayofyear
```

```python
df1['QUARTER'] = pd.PeriodIndex(df1['GAME_DATE_EST'], freq='Q')
```

```python
for team in list(set(df1['NAME_HOME'])):
    date_dict[team] = pd.Timestamp('2004-06-15 00:00:00')
```

```python
df1.drop(['day_after_last_game'],axis = 1, inplace=True)
```

```python
df1['day_after_last_game_home'] = 0
df1['day_after_last_game_away'] = 0
```

```python
df1.sort_values(by=['GAME_DATE_EST'],ascending=False,inplace = True)
```

```python
for row in df1[::-1].itertuples():
    date = row.GAME_DATE_EST
    interval_home = (date - date_dict[row.NAME_HOME]).days
    interval_away = (date - date_dict[row.NAME_AWAY]).days
    df1.at[row.Index, 'day_after_last_game_home'] = interval_home
    df1.at[row.Index, 'day_after_last_game_away'] = interval_away
    date_dict[row.NAME_HOME] = date
    date_dict[row.NAME_AWAY] = date
```

```python
rank = pd.read_csv('newdat/rankings.csv')
```

```python
rank[rank['SEASON']==2003]
```

```python
df1['block_home'] = 0
df1['block_away'] = 0
```

```python
for row in df1[::-1].itertuples():
    season = row.SEASON
    #print(season)
    home_team = row.TEAM_ID_home
    away_team = row.TEAM_ID_away
    df1.at[row.Index, 'block_home'] = int(rank.loc[(rank['SEASON'] == season-1) & (rank['TEAM_ID'] == home_team)]['block'])
    df1.at[row.Index, 'block_away'] = int(rank.loc[(rank['SEASON'] == season-1) & (rank['TEAM_ID'] == away_team)]['block'])
```

```python
df1['type_home'] = 'NA'
df1['type_away'] = 'NA'
for row in df1[::-1].itertuples():
    a = row.block_home
    b = row.block_away
    if a == 1 and b == 1:
        df1.at[row.Index, 'type_home'] = 'SS'
        df1.at[row.Index, 'type_away'] = 'SS'
    elif a == 1 and b == 2:
        df1.at[row.Index, 'type_home'] = 'SM'
        df1.at[row.Index, 'type_away'] = 'MS'
    elif a == 1 and b == 3:
        df1.at[row.Index, 'type_home'] = 'SW'
        df1.at[row.Index, 'type_away'] = 'WS'
    elif a == 2 and b == 1:
        df1.at[row.Index, 'type_home'] = 'MS'
        df1.at[row.Index, 'type_away'] = 'SM'
    elif a == 2 and b == 2:
        df1.at[row.Index, 'type_home'] = 'MM'
        df1.at[row.Index, 'type_away'] = 'MM'
    elif a == 2 and b == 3:
        df1.at[row.Index, 'type_home'] = 'MW'
        df1.at[row.Index, 'type_away'] = 'WM'
    elif a == 3 and b == 1:
        df1.at[row.Index, 'type_home'] = 'WS'
        df1.at[row.Index, 'type_away'] = 'SW'
    elif a == 3 and b == 2:
        df1.at[row.Index, 'type_home'] = 'WM'
        df1.at[row.Index, 'type_away'] = 'MW'
    elif a == 3 and b == 3:
        df1.at[row.Index, 'type_home'] = 'WW'
        df1.at[row.Index, 'type_away'] = 'WW'
```

```python
df1.to_csv('newdat/games_with_block.csv',index=False)
```
