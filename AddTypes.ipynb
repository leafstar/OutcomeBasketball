```python
import pandas as pd
```

```python
rank = pd.read_csv('newdat/rankings.csv')
```

```python
rank
```

```python
df1 = pd.read_csv('newdat/games_with_salary.csv')
```

```python
df1['block_home_2'] = 0 # 2 levels
df1['block_away_2'] = 0 # 2 levels
df1['block_home_3'] = 0 # 5 levels
df1['block_away_3'] = 0 # 5 levels
```

```python
for row in df1[::-1].itertuples():
    season = row.SEASON
    #print(season)
    home_team = row.TEAM_ID_home
    away_team = row.TEAM_ID_away
    df1.at[row.Index, 'block_home_2'] = int(rank.loc[(rank['SEASON'] == season-1) & (rank['TEAM_ID'] == home_team)]['block2'])
    df1.at[row.Index, 'block_away_2'] = int(rank.loc[(rank['SEASON'] == season-1) & (rank['TEAM_ID'] == away_team)]['block2'])
    df1.at[row.Index, 'block_home_3'] = int(rank.loc[(rank['SEASON'] == season-1) & (rank['TEAM_ID'] == home_team)]['block3'])
    df1.at[row.Index, 'block_away_3'] = int(rank.loc[(rank['SEASON'] == season-1) & (rank['TEAM_ID'] == away_team)]['block3'])
```

```python
df1
```

```python
df1['type_home_2'] = 'NA'
df1['type_away_2'] = 'NA'
df1['type_home_3'] = 'NA'
df1['type_away_3'] = 'NA'
for row in df1[::-1].itertuples():
    a = row.block_home_2
    b = row.block_away_2
    df1.at[row.Index, 'type_home_2'] = str(a)+'v'+str(b)
    df1.at[row.Index, 'type_away_2'] = str(b)+'v'+str(a)
    
    a = row.block_home_3
    b = row.block_away_3
    df1.at[row.Index, 'type_home_3'] = str(a)+'v'+str(b)
    df1.at[row.Index, 'type_away_3'] = str(b)+'v'+str(a)
    
```
