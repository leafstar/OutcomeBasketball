```python
import pandas as pd
df1 = pd.read_csv('newdat/games_extended.csv')
```

```python

path = './nba_data'
teams = pd.read_csv(path + '/teams.csv')
```

```python
from math import sin,asin, cos, sqrt, atan2, radians,pi
def distance(lat1, lon1, lat2, lon2):
    p = pi/180
    a = 0.5 - cos((lat2-lat1)*p)/2 + cos(lat1*p) * cos(lat2*p) * (1-cos((lon2-lon1)*p))/2
    return 12742 * asin(sqrt(a)) #2*R*asin...
```

```python
teams.drop(["LEAGUE_ID", "MIN_YEAR", "MAX_YEAR", "ABBREVIATION", "YEARFOUNDED", "CITY", "ARENA", "ARENACAPACITY", "OWNER", "GENERALMANAGER", "DLEAGUEAFFILIATION","HEADCOACH"], axis = 1, inplace = True)
```

```python
merged1 = pd.merge(df1, teams, left_on = "TEAM_ID_home", right_on='TEAM_ID', how='left')
```

```python
merged1.drop(["TEAM_ID","NICKNAME"],axis = 1,inplace = True)
merged1.rename(columns={"longitude": "longi_HOME", "latitude": "lati_HOME"}, inplace = True)
```

```python
merged2 = pd.merge(merged1, teams, left_on = "TEAM_ID_away", right_on='TEAM_ID', how='left')
merged2.drop(["TEAM_ID","NICKNAME"],axis = 1,inplace = True)
merged2.rename(columns={"longitude": "longi_AWAY", "latitude": "lati_AWAY"}, inplace = True)
```

```python
merged2['distance_for_away'] = merged2[['longi_HOME', 'lati_HOME', 'longi_AWAY', 'lati_AWAY']].apply(lambda x: distance(x['lati_HOME'],x['longi_HOME'],  x['lati_AWAY'],x['longi_AWAY']), axis=1)
```
